```python
def generate_gemini_prompt(code: str, lang: str = "Python") -> str:
    return f"""
You are a helpful AI assistant. Refactor the following {lang} code:

### GOALS:
- Improve naming (variables/functions)
- Split large functions if needed
- Optimize performance (loops, logic)
- Follow clean code practices

### ORIGINAL CODE ({lang}):
{code}

### REFACTORED CODE ({lang}):
"""
```

```python
def refactor_with_gemini(code: str, lang: str = "Python"):
    prompt = generate_gemini_prompt(code, lang)

    model = genai.GenerativeModel("gemini-pro")  # or use "gemini-pro"
    response = model.generate_content(prompt)

    return response.text.strip()
```
